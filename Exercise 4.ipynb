```python
import matplotlib.pyplot as plt
from astropy.io import fits
hdu=fits.open('./Skyserver_Radial9_19_2017 7_38_23 AM.fits')
```

```python
datas=hdu[1].data
datas
```

```python
u=datas.field(9)
g=datas.field(10)
r=datas.field(11)
```

```python
u_g=u-g
g_r=g-r
```

```python
fig, ax = plt.subplots()
plt.figure(1)
fig.set_size_inches(20, 15)
legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')
#legend.get_frame().set_facecolor('#00FFCC')
plt.xlabel('u-g[mag]',fontsize=20)
plt.ylabel('g-r[mag]',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(u_g,g_r,'bo',10)
plt.axis([1,2.5,0,2])
plt.show()
```

# Description:
## The pattern seems to be that "g-r" magnitude and "u-g" magnitude are positively related.
