```python
#import package
import matplotlib.pyplot as plt
from astropy.io import fits
hdu=fits.open('Skyserver_Radial9_19_2017 8_52_39 AM.fits')
```

```python
#get data units
datas=hdu[1].data
```

```python
#show datas
datas.columns
```

```python
datas
```

```python
#get data
u=datas.field(3)
g=datas.field(4)
r=datas.field(5)
```

```python
u_g=u-g
g_r=g-r
```

```python
fig, ax = plt.subplots()
plt.figure(1)
fig.set_size_inches(20, 15)
#legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')
#legend.get_frame().set_facecolor('#00FFCC')
plt.xlabel('u-g[mag]',fontsize=20)
plt.ylabel('g-r[mag]',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(u_g,g_r,'bo',10)
#plt.axis([1,2.5,0,2])
plt.show()
```

#### Answer:
##### Elliptical Galaxies
    Elliptical galaxies, generally have very little gas and dust. Because gas and dust are found in the clouds that are the birthplaces of stars, we should expect to see very few young stars in elliptical galaxies. In fact, elliptical galaxies contain primarily old, red stars. 
    So those redder galaxies on the color-color map are Elliptical Galaxies.
#### Sprial Galaxies
    Spiral galaxies usually have two distinct regions. 
    The disk of the galaxy contains the spiral arms. The disk is a region of star formation and has a great deal of gas and dust,has large SFR. It is dominated by young, blue stars,thes stars may be O,B stars,so they are very bright and blue!
    The central bulge is devoid of gas and dust, the bulge is composed primarily of Population II stars.
    So those bluer galaxies on the color-color map are Spiral Galaxies.
