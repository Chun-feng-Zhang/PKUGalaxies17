```python
#read file 
def star(file):
    f=open(file,'r')
    wavlen=[]
    flux=[]
    
    header1 = f.readline()
    header2 = f.readline()
    header2 = f.readline()
    for line in f:
        line=line.strip()
        columns=line.split()
        sou1=float(columns[0])
        sou2=float(columns[1])
        wavlen.append(sou1)
        flux.append(sou2)
    f.close()
    return wavlen,flux
```

```python

#plot spectra
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np
import matplotlib as mpl

mpl.style.use('seaborn')
plt.figure(3)
fig=plt.figure(figsize=(40,30))

def plotstar(subplot,file,title):
    ax1=plt.subplot(subplot)
    ax1.set_xlabel('Wavelength [ngstrms]',fontsize=20)
    ax1.set_ylabel('Flux [erg/cm2/s/]',fontsize=20)
    ax1.set_xlim(3800,9200)
    ax1.set_title(title,fontsize=20)
    ax1.set_ylim(0,5)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.subplots_adjust(top=0.92, bottom=0.06, hspace=0.25,wspace=0.35)
    star1=star(file)
    plt.plot(star1[0],star1[1],'C0',linewidth=1)
    

```

```python
# get data
#define function
#read file 
def starsyn(file,weight):
    f=open(file,'r')
    wavlen=[]
    flux=[]
    
    header1 = f.readline()
    header2 = f.readline()
    header2 = f.readline()
    for line in f:
        line=line.strip()
        columns=line.split()
        sou1=float(columns[0])
        sou2=float(columns[1])
        wavlen.append(sou1)
        flux.append(sou2)
    f.close()
    flux=weight*np.array(flux)
    return wavlen,flux
```

```python
#Through files
import os
import os.path
def getdata(dr,i):
    
    
    #dr='./o-mstar/Ostar/'
    dataresult=[]
    for parent,dirnames,filenames in os.walk(dr):
    #print(parent)
    #print('\n')
    #print(dirnames)
        print('ok')
    #print(filenames)
    for filename in filenames:
        file=os.path.join(parent,filename)
        dataresult.append(starsyn(file,weight[i]))# 3 arrays in specweight array,means 5v,8iii,9v=specweight[0],[1],[2];;each array has two column,including wavelength and flux
    return dataresult
```

```python
#get all star spectrum
odata=getdata('./o-mstar/Ostar/',0)#3 stars   ---len(odata)=3
bdata=getdata('./o-mstar/Bstar/',1)#19 stars
adata=getdata('./o-mstar/Astar/',2)#
fdata=getdata('./o-mstar/Fstar/',3)
gdata=getdata('./o-mstar/Gstar/',4)
kdata=getdata('./o-mstar/Kstar/',5)
mdata=getdata('./o-mstar/Mstar/',6)
```

```python
#spectype =O,B,A,F,G,K,M and '',input O means ignore O stars,'' means ignore nothing 
def subtract(spectype):
    
    #combine all the star spectrum
    sspspec=np.zeros(4771,float)
        
    for each in odata:
        if spectype=='O':
            continue
        sspspec+=each[1]
    for each in bdata:
        if spectype=='B':
            continue
        sspspec+=each[1]
    for each in adata:
        if spectype=='A':
            continue
        sspspec+=each[1]
    for each in fdata:
        if spectype=='F':
            continue
        sspspec+=each[1]
    for each in gdata:
        if spectype=='G':
            continue
        sspspec+=each[1]
    for each in kdata:
        if spectype=='K':
            continue
        sspspec+=each[1]
    for each in mdata:
        if spectype=='M':
            continue
        sspspec+=each[1]
    sspwav=each[0]
    return sspwav,sspspec
```

```python
mpl.style.use('seaborn')
plt.figure(5)
fig=plt.figure(figsize=(20,15))
ax1=plt.subplot(111)
ax1.set_xlabel('Wavelength [ngstrms]',fontsize=20)
ax1.set_ylabel('Flux [erg/cm2/s/]',fontsize=20)
ax1.set_xlim(3800,25000)
#ax1.set_ylim(0,)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax1.set_title('Spectrum of SSP',fontsize=20)
result=subtract('')#getdata from subtrct function,we get sspwav and sspspec
sspwav=result[0]
sspspec=result[1]
plt.plot(sspwav,sspspec,'C0',linewidth=1)

plt.show()
```

##### Three,
calculate ssp flux of u,g,r bands and color respectively. 

```python
def getcolor(spectype):
#spectype =O,B,A...M,means the ssp color after ignoring O,B...stars,when spectype='',which means ignore nothing   

    #color u-g,g-r
    #fwhm mean full width half maximum ,unit is ngstrms
    ufwhm=660#central wavlength  3551
    gfwhm=1000#cetral wavlength   4686
    rfwhm=1380#central wavelength 6166
    #find index of 3551 in sspwav and then use the return value to determine the flux in sspspec
    result=subtract(spectype)
    sspwav=result[0]
    sspspec=result[1]
    indexu=sspwav.index(3550.0)
    indexg=sspwav.index(4685.0)
    indexr=sspwav.index(6165.0)

    uflux=sspspec[indexu]
    gflux=sspspec[indexg]
    rflux=sspspec[indexr]

    u_g=-2.5*np.log10(ufwhm*uflux/(gfwhm*gflux))#u-g color
    g_r=-2.5*np.log10(gfwhm*gflux/(rfwhm*rflux))#g-r color
    return u_g,g_r
```

```python
##spectype =O,B,A,F,G,K,M and '',input O means ignore O stars,'' means ignore nothing 
def color_evolution(spectype,label):
    colorres=getcolor(spectype)#get u-g,g-r form getcolor function
    u_g=colorres[0]
    g_r=colorres[1]
    
    
    mpl.style.use('seaborn')
    #plt.figure(6)
    #fig=plt.figure(figsize=(20,15))
    #ax1=plt.subplot(111)
    ax1.set_xlabel('u-g',fontsize=20)
    ax1.set_ylabel('g-r',fontsize=20)

    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    ax1.set_title('ssp color evolution',fontsize=20)
    ax1.scatter(u_g,g_r,60,c='C0')
    plt.text(u_g,g_r+0.005,label,fontweight='bold',fontsize=20)

    
```

```python
plt.figure(6)
fig=plt.figure(figsize=(20,15))
ax1=plt.subplot(111)
color_evolution('',0)#all stars in ssp are alive----t0
color_evolution('O',1)#all O stars die--------------t1
color_evolution('B',2)#all B stars die--------------t2
color_evolution('A',3)#all A  stars die-------------t3
color_evolution('F',4)#...--------------------------t4
color_evolution('G',5)#...--------------------------t5
color_evolution('K',6)#..---------------------------t6
color_evolution('M',6)#..---------------------------t7
plt.show()
```
